```python
tri_data_clean = industry_reduced[["DOCUMENT_NUMBER",'FACILITY_NAME','PARENT_COMPANY_NAME','STATE','LATITUDE', 'LONGITUDE', 
                             'FEDERAL_FACILITY','TRIBAL_LAND','INDUSTRY_SECTOR_CODE', 'INDUSTRY_SECTOR',
                             'SRS_ID', 'CHEMICAL', 'METAL_CATEGORY', 'CARCINOGEN','ON_SITE_RELEASE_TOTAL',
                             'OFF_SITE_RELEASE_TOTAL','OFF_SITE_RECYCLED_TOTAL','ONE_TIME_RELEASE',
                             'UNIT_OF_MEASURE']]

cleaned_all = pd.DataFrame(tri_data_clean)
```

```python
cleaned_all["REGION"] = ""
cleaned_all
```

```python
states = cleaned_all["STATE"]

west = ['WA','OR','MT','ID','WY','CO','UT','NV','CA','HI','AK']
southwest= ['AZ','NM','OK','TX']
midwest = ['ND','SD','NE','KS','MN','IA','MO','WI','IL','MI','IN','OH']
southeast = ['AR','LA','MS','AL','GA','FL','SC','NC','TN','KY','VA','DC','WV']
northeast = ['MD','DE','NJ','PA','RI','NY','CT','MA','NH','VT','ME']

region = []
for state in states:
    if state in west:
        region.append("WEST REGION")
    elif state in southwest:
        region.append("SOUTHWEST REGION")
    elif state in midwest:
        region.append("MIDWEST REGION")
    elif state in southeast:
        region.append("SOUTHEAST REGION")
    else:
        region.append("NORTHEAST REGION")
```

```python
cleaned_all["REGION"] = region
clean_data = cleaned_all.copy()
clean_data
```

```python
clean_data["TRIBAL_LAND"].fillna("NO", inplace = True)
clean_data["PARENT_COMPANY_NAME"].fillna("NONE", inplace=True)
clean_data
```

```python
i = 0
test_list = []
while i < len(clean_data["DOCUMENT_NUMBER"]):
    tribes = clean_data["TRIBAL_LAND"].iloc[i]
    if tribes == "NO":
        test_list.append(clean_data["REGION"].iloc[i])
        i = i + 1
    else:
        test_list.append("TRIBAL LAND")
        i = i + 1

clean_data["REGION"] = test_list
```

```python
tribal = clean_data.loc[(clean_data["TRIBAL_LAND"] != "NO"), :]
tribal_list = tribal[["DOCUMENT_NUMBER", "TRIBAL_LAND"]]
tribal_list

tribal_list.to_csv("Output/tribes.csv", encoding="UTF-8", index=False)
```

```python
industry = clean_data[["INDUSTRY_SECTOR_CODE", "INDUSTRY_SECTOR"]]
industry_list = industry.groupby(["INDUSTRY_SECTOR_CODE"]).first()
industry_list.index = industry_list.index.astype("int64")
industry_list["INDUSTRY_SECTOR_CODE"] = industry_list.index
industry_list = industry_list[["INDUSTRY_SECTOR_CODE", "INDUSTRY_SECTOR"]]
industry_list

industry_list.to_csv("Output/industry.csv", encoding="UTF-8", index=False)
```

```python
chemical = clean_data[["SRS_ID", "CHEMICAL"]]
chemical_list = chemical.groupby(["SRS_ID"]).first()
chemical_list.index = chemical_list.index.astype("int64")
chemical_list["SRS_ID"] = chemical_list.index
chemical_list = chemical_list[["SRS_ID", "CHEMICAL"]]
chemical_list

chemical_list.to_csv("Output/chemicals.csv", encoding="UTF-8", index=False)
```

```python
clean_dict = clean_data[["DOCUMENT_NUMBER",'FACILITY_NAME','PARENT_COMPANY_NAME','STATE','REGION', 'LATITUDE', 'LONGITUDE', 
                             'FEDERAL_FACILITY','TRIBAL_LAND','INDUSTRY_SECTOR_CODE', 'INDUSTRY_SECTOR',
                             'SRS_ID', 'CHEMICAL', 'METAL_CATEGORY', 'CARCINOGEN','ON_SITE_RELEASE_TOTAL',
                             'OFF_SITE_RELEASE_TOTAL','OFF_SITE_RECYCLED_TOTAL','ONE_TIME_RELEASE',
                             'UNIT_OF_MEASURE']]

clean_dict.head()
```

```python
region_list = []
chart_dict = pd.DataFrame()
region_names = sorted(clean_dict["REGION"].unique())
#chart_dict["Region_names"] = region_names
```

```python
data_list = {}

for item in region_names:
    data_dict = {}

    pick_region = clean_dict.loc[(clean_dict["REGION"] == item), :]
    chemicals = pick_region.groupby(["CHEMICAL"]).sum()
    chemical_sums = list(chemicals["ON_SITE_RELEASE_TOTAL"])

    industries = pick_region.groupby(["INDUSTRY_SECTOR"]).sum()
    industry_list = list(industries["ON_SITE_RELEASE_TOTAL"])

    states = pick_region.groupby(["STATE"]).sum()
    state_list = list(states["ON_SITE_RELEASE_TOTAL"])

    data_dict["Region"] = item
    data_dict["Facility_name"] = list(pick_region["FACILITY_NAME"].unique())
    data_dict["Parent_co_name"] = list(pick_region["PARENT_COMPANY_NAME"].unique())
    data_dict["State"] = list(states.index)
    data_dict["State_sum"] = [item/2000 for item in state_list]
    data_dict["SRS_ID"] = list(pick_region["SRS_ID"].unique())
    data_dict["Chemical_list"] = list(chemicals.index)
    data_dict["Chemical_sums"] = [item/2000 for item in chemical_sums]
    data_dict["Industry_code"] = list(pick_region["INDUSTRY_SECTOR_CODE"].unique())
    data_dict["Industry"] = list(industries.index)
    data_dict["Industry_sums"] = [item/2000 for item in industry_list]
    data_dict["Region_count"] = len(pick_region["CARCINOGEN"])
    data_dict["Carcinogen_count"] = len(pick_region.loc[(pick_region["CARCINOGEN"] == "YES"), :])
    data_dict["On_site_region"] = (pick_region["ON_SITE_RELEASE_TOTAL"].sum())/2000
    data_dict["Off_site"] = (pick_region["OFF_SITE_RELEASE_TOTAL"].sum())/2000
    data_dict["Recycling"] = (pick_region["OFF_SITE_RECYCLED_TOTAL"].sum())/2000

    region_list.append(data_dict)
```

```python
chart_dict["Region_data"] = region_list
chart_dict
```
