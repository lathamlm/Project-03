```python
#IMPORT DEPENDENCIES

import matplotlib.pyplot as plt
import pandas as pd
from pathlib import Path
```

```python
#OPEN/READ FILE
file = Path("Resources/2021_us.csv")

tri_df = pd.read_csv(file, encoding='UTF-8')
tri_df.head()
```

```python
tri_columns = tri_df[["4. FACILITY NAME", "11. TRIBE", "12. LATITUDE","13. LONGITUDE", "15. PARENT CO NAME", "18. FEDERAL FACILITY", "19. INDUSTRY SECTOR CODE", 
                      "20. INDUSTRY SECTOR", "34. CHEMICAL", "42. METAL CATEGORY", "43. CARCINOGEN", "47. UNIT OF MEASURE",
                      "62. ON-SITE RELEASE TOTAL", "85. OFF-SITE RELEASE TOTAL", "91. OFF-SITE RECYCLED TOTAL", "117. 8.8 - ONE-TIME RELEASE","8. ST", "33. DOC_CTRL_NUM",
                      "38. SRS ID"]]
tri_columns
```

```python
tri_columns_rename = tri_columns.rename(columns={"4. FACILITY NAME":"FACILITY NAME", "11. TRIBE":"TRIBAL LAND", "12. LATITUDE":"LATITUDE","13. LONGITUDE":"LONGITUDE", 
                                                "15. PARENT CO NAME":"PARENT COMPANY NAME", "18. FEDERAL FACILITY":"FEDERAL FACILITY", "19. INDUSTRY SECTOR CODE":"INDUSTRY SECTOR CODE", 
                                                "20. INDUSTRY SECTOR":"INDUSTRY SECTOR", "34. CHEMICAL":"CHEMICAL", "42. METAL CATEGORY":"METAL CATEGORY", "43. CARCINOGEN":"CARCINOGEN",
                                                "47. UNIT OF MEASURE":"UNIT OF MEASURE","62. ON-SITE RELEASE TOTAL":"ON-SITE RELEASE TOTAL", "85. OFF-SITE RELEASE TOTAL":"OFF-SITE RELEASE TOTAL",
                                                "91. OFF-SITE RECYCLED TOTAL":"OFF-SITE RECYCLED TOTAL", "117. 8.8 - ONE-TIME RELEASE":"ONE-TIME RELEASE",
                                                "8. ST": "STATE", "33. DOC_CTRL_NUM": "DOCUMENT NUMBER", "38. SRS ID":"SRS ID"})
tri_columns_rename.head()
```

```python
tri_pounds = tri_columns_rename.loc[tri_columns_rename["UNIT OF MEASURE"] == 'Pounds']
tri_pounds.head(20)
```

```python
tri_pounds.columns
```

```python
tri_data_clean = tri_pounds[["DOCUMENT NUMBER",'FACILITY NAME','PARENT COMPANY NAME','STATE','LATITUDE', 'LONGITUDE', 
                             'FEDERAL FACILITY', 'TRIBAL LAND','INDUSTRY SECTOR','INDUSTRY SECTOR CODE', 
                             'CHEMICAL','SRS ID', 'METAL CATEGORY', 'CARCINOGEN','ON-SITE RELEASE TOTAL',
                             'OFF-SITE RELEASE TOTAL','OFF-SITE RECYCLED TOTAL','ONE-TIME RELEASE',
                             'UNIT OF MEASURE']]
tri_data_clean
```

```python
tri_data_clean.to_csv("tri_data_clean.csv", encoding="utf-8", index=False)
```

```python
industry = tri_data_clean[["INDUSTRY SECTOR CODE", 
                      "INDUSTRY SECTOR"]]
industry.groupby(["INDUSTRY SECTOR CODE"]).first()
industry.to_csv("industry.csv", encoding="utf-8", index=False)
```

```python
chemicals = tri_data_clean[["CHEMICAL","SRS ID"]]
chemicals.to_csv("chemicals.csv", encoding="utf-8", index=False)
```

```python
chemical_count = chemicals["CHEMICAL"].value_counts()
#chemical_count.head(25)
chemical_count.replace({"Lead  And Lead Compounds":"Lead And Lead Compounds"})
chemical_count.head(25)
```

```python
tribal_land = tri_data_clean[["TRIBAL LAND"]]
tribal_land.to_csv("tribal_land.csv", encoding="utf-8", index=False)
```

```python
industry_cleaned = industry.loc[(industry["INDUSTRY SECTOR"] == "Chemicals") | (industry["INDUSTRY SECTOR"] == "Fabricated Metals") | (industry["INDUSTRY SECTOR"] == "Primary Metals") | (industry["INDUSTRY SECTOR"] == "Nonmetallic Mineral Product") |
               (industry["INDUSTRY SECTOR"] == "Transportation Equipment") | (industry["INDUSTRY SECTOR"] == "Petroleum Bulk Terminals") | (industry["INDUSTRY SECTOR"] == "Petroleum") | (industry["INDUSTRY SECTOR"] == "Food") |
              (industry["INDUSTRY SECTOR"] == "Machinery") | (industry["INDUSTRY SECTOR"] == "Plastics and Rubber") | (industry["INDUSTRY SECTOR"] == "Electric Utilities") | (industry["INDUSTRY SECTOR"] == "Chemical Wholesalers") |
               (industry["INDUSTRY SECTOR"] == "Other") | (industry["INDUSTRY SECTOR"] == "Paper") | (industry["INDUSTRY SECTOR"] == "Hazardous Waste") | (industry["INDUSTRY SECTOR"] == "Electrical Equipment") |
               (industry["INDUSTRY SECTOR"] == "Computers and Electronic Products") | (industry["INDUSTRY SECTOR"] == "Wood Products"), :]
industry_cleaned.drop_duplicates(subset="INDUSTRY SECTOR CODE",
                     keep=False)
industry_cleaned.head(30)
```

```python
chemical_cleaned = (chemicals.loc[(chemicals["CHEMICAL"] == "Lead") | (chemicals["CHEMICAL"] == "Zinc compounds") | (chemicals["CHEMICAL"] == "Nickel") | (chemicals["CHEMICAL"] == "Chromium") | (chemicals["CHEMICAL"] == "Copper") |
                   (chemicals["CHEMICAL"] == "Nitrate compounds (water dissociable; reportable only when in aqueous solution)") | (chemicals["CHEMICAL"] == "Ammonia") | (chemicals["CHEMICAL"] == "Manganese") |
                   (chemicals["CHEMICAL"] == "Methanol") | (chemicals["CHEMICAL"] == "Toluene") | (chemicals["CHEMICAL"] == "Lead  And Lead Compounds") | (chemicals["CHEMICAL"] == "Xylene (mixed isomers)") | (chemicals["CHEMICAL"] == "Lead compounds") |
                   (chemicals["CHEMICAL"] == "Nitric acid") | (chemicals["CHEMICAL"] == "Polycyclic aromatic compounds") | (chemicals["CHEMICAL"] == "Certain glycol ethers") | (chemicals["CHEMICAL"] == "Diisocyanates") |
                   (chemicals["CHEMICAL"] == "Ethylene glycol") | (chemicals["CHEMICAL"] == "Ethylbenzene") | (chemicals["CHEMICAL"] == "Naphthalene") | (chemicals["CHEMICAL"] == "1,2,4-Trimethylbenzene") | (chemicals["CHEMICAL"] == "n-Hexane") |
                   (chemicals["CHEMICAL"] == "Styrene") | (chemicals["CHEMICAL"] == "Benzo[g,h,i]perylene") | (chemicals["CHEMICAL"] == "Manganese compounds"), :])
chemical_cleaned.drop_duplicates(subset="CHEMICAL",
                     keep=False)
chemical_cleaned.head(25)
```

```python
industry_cleaned.to_csv("industry_cleaned.csv", encoding="utf-8", index=False)
chemical_cleaned.to_csv("chemical_cleaned.csv", encoding="utf-8", index=False)
```
